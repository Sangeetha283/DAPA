```python
import matplotlib.pyplot as plt

manufactures = ['samsung', 'apple', 'huawei', 'xiaomi', 'others']
market_share = [30,25,18,12,15]

plt.pie(market_share, labels=manufactures)

plt.title('smart phone market share')
plt.show()
```

```python
import matplotlib.pyplot as plt
income_source = ["salary", "freelance", "investment", "rental", "others"]
monthly_income = [5000,1500,1000,600,400]
plt.pie(monthly_income, labels=income_source)

plt.title('')
plt.show()
```

```python
import matplotlib.pyplot as plt

segments = ['product A', 'product B', 'services', 'licensing']
revenue_percentage = [45,25,15,15]
plt.pie(revenue_percentage, labels=segments)

plt.title('')
plt.show()
```

```python
categories = ['rent', 'food', 'transport', 'entertainment','saving']
values = [35,25,15,10,15]
plt.pie(values, labels=categories)

plt.title('')
plt.show()

```

```python
SCATTER PLOT: To show relationship between 2 variables
```

```python
study_hours = [2,3,1,4,3,5,2,6,5,7,20,10,50,60,70]
exam_scores = [65,75,60,80,70,85,70,90,88,92,100,50,60,100,90]

plt.scatter(study_hours, exam_scores, label="student data")
plt.xlabel("study hours")
plt.ylabel("exam scores")
plt.title("study hours vs exam scores")
plt.show()
```
